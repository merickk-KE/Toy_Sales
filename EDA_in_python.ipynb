### Multivariate analysis


```python
# looking for how symmetrical the data is using the skew function
df.skew(numeric_only=True)
```

```python
# looking for outliers in our data using kurtosis(<3 means no outliers and vice versa)
df.kurtosis(numeric_only=True)
```

```python

plt.hist(df['target_age_group'], bins =10)
plt.show()
```

```python
features = df.select_dtypes(include='number')
features.plot(kind = 'box', subplots=True, layout=(3, 3), sharex=False, figsize=(10, 10), grid=True)
plt.tight_layout()
plt.show()
```

```python
# Correlations
plt.figure(figsize =(8, 8))
heatmap = sns.heatmap(df.corr(numeric_only=True), vmin=-1, vmax=1, annot=True)

heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(data=features)
plt.show()
```
