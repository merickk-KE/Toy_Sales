### city column 

```python
df['city'].unique()
```

```python
# cleaning the city column 
df['city'] = df['city'].str.strip()
```

```python
import duckdb
import pandas as pd

df = pd.read_csv(r"C:\Users\lenovo\Downloads\indian toys\cleaned_indian_toy_market_sales_data.csv")

query = """
SELECT DISTINCT city, region
FROM df
WHERE region = 'West'
GROUP BY city, region

-- LIMIT 30
"""

result = duckdb.query(query).to_df()
result
```

```python
regions_dict = {
    'South':['Kochi', 'Mysore','Coimbatore','Bengaluru', 'Hyderabad', 'Chennai' ], 
    'North' : ['Delhi', 'Noida', 'Chandigarh', 'Gurugram','Jaipur', 'Lucknow' ],
    'East' : ['Patna', 'Kolkata', 'Guwahati', 'Ranchi', 'Bhubaneswar'],
    'West' : ['Indore', 'Mumbai', 'Pune', 'Surat', 'Ahmedabad', 'Nagpur']
}


# Reverse mapping: city -> region
regions_cities = {
    city : region
    for region, cities in regions_dict.items()
    for city in cities
}

# Fill only rows where region is NaN
mask = df['region'].isna()

df.loc[mask, 'region'] = (
    df.loc[mask, 'city']
      .astype(str)
      .map(regions_cities)
)
```

#### cleaning the date column 

```python
print(df['transaction_date'].head(10))
```

```python
# replacing / with - to standardize the transaction_date column 
df['transaction_date'] = df['transaction_date'].str.replace('/', '-')
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format = 'mixed', dayfirst=True)
```
