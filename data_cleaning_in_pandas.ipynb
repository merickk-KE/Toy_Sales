## Data Cleaning using Pandas 

```python
# importing the required packages 
import pandas as pd
```

```python

df = pd.read_csv(r"C:\Users\lenovo\Downloads\indian toys\indian_toy_market_sales_data.csv")
df.head()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
# changing column names to snake case
df.columns = df.columns.str.lower()
```

```python
df.head()
```

```python
df['safety_certified'] = df['safety_certified'].map({'TRUE': 'Yes',
                                                     'yes': 'Yes', 
                                                     'No':'No', 
                                                     'N':'No', 
                                                     'Y':'Yes', 
                                                     'no': 'No', 
                                                     'Yes':'Yes', 
                                                     'FALSE': 'No'})
print(f'safety_certified column cleaned.')
```

```python
df['safety_certified'].unique()
```

### Brand column

```python
# cleaning the brand column 
df['brand'].unique()
```

```python
# trimming the right and left spaces 
df['brand'] = (df['brand'].str.strip())
```

```python
df['brand'].unique()
```

```python
df['brand'] = df['brand'].map({'OKPlay' : 'OK Play',
                       'Funskool India' : 'Funskool', 
                       'Lego' : 'Lego',
                       'Fisher Price': 'Fisher Price',
                       'HASBRO':'Hasbro', 
                       'Nerf' : 'Nerf',
                       'Funskool' : 'Funskool',
                       'OK Play' : 'OK Play',
                       'Hot Wheels' : 'Hot Wheels',
                       'Barbie' : 'Barbie',
                       'Hasbro' : 'Hasbro', 
                       'Skillmatics' : 'Skillmatics',
                       'Toy Zone' : 'Toy Zone',
                       'Playshifu' : 'Playshifu',
                       'Mattel' : 'Mattel',
                       'Smartivity': 'Smartivity',
                       'HotWheels' : 'Hot Wheels', 
                       'Toyzone' : 'Toy Zone',  
                       'NERF' : 'Nerf', 
                       'LEGO' : 'Lego',
                       'BARBIE' :'Barbie', 
                       'Fisher-Price' : 'Fisher Price', 
                       'MATTEL' : 'Mattel',
                       })

```

```python
df['brand'].unique()
```

### city column 

```python
df['city'].unique()
```

```python
# cleaning the city column 
df['city'] = df['city'].str.strip()
```

```python
city_rename = {
    'Calcutta': 'Kolkata',
    'Bombay': 'Mumbai',
    'Bangalore': 'Bengaluru'
}
df['city'] = df['city'].replace(city_rename)
```

#### sales_channel column

```python
df['sales_channel'].unique()
```

```python
# cleaning the sales_channel column 
channel_replace = {
    'Hamleys':'Hamleys Store',
    'amazon.in': 'Amazon',
    'flipkart' : 'Flipkart',
    'Website' : 'Brand Website'
}
df['sales_channel'] = df['sales_channel'].replace(channel_replace)
```

#### target_age_group column 

```python
df['target_age_group'].unique()
```

```python
age_group_replace = {
    '0-2' : '0-2 years',
    '3-5' : '3-5 years', 
    '6-8' : '6-8 years',
    '9-12' : '9-12 years',
    '13+' : '13+ years' 
}
df['target_age_group']= df['target_age_group'].replace(age_group_replace)
```

#### Quantity column 

```python
df['quantity'].unique()
```

```python
df['quantity'] = df['quantity'].abs()
```

####  returned column

```python
df['returned'].unique()
```

```python
returned_std = {
    'yes' : 'Yes',
    'Y' : 'Yes',
    'N' : 'No',
    'no' : 'No'
}
df['returned'] = df['returned'].replace(returned_std)
```

#### cleaning the date column 

```python
print(df['transaction_date'].head(10))
```

```python
# replacing / with - to standardize the transaction_date column 
df['transaction_date'] = df['transaction_date'].str.replace('/', '-')
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format = 'mixed', dayfirst=True)
```

```python
# printing duplicates from the data 
print(df[df.duplicated()])
```
